```python
import numpy as np
import pandas as pd
import torch
import evaluate

from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    DataCollatorForSeq2Seq,
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer,
    TrainerCallback
)

dataset = load_dataset("abisee/cnn_dailymail", "3.0.0")  # parquet-friendly
```

```python
model_checkpoint = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

max_input_length = 512
max_target_length = 128

def preprocess_function(examples):
    inputs = ["summarize: " + x for x in examples["article"]]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    labels = tokenizer(text_target=examples["highlights"], max_length=max_target_length, truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_data = dataset["train"].select(range(10000))
val_data   = dataset["validation"].select(range(1000))

tokenized_train = train_data.map(preprocess_function, batched=True, remove_columns=train_data.column_names)
tokenized_val   = val_data.map(preprocess_function, batched=True, remove_columns=val_data.column_names)
```

```python
rouge = evaluate.load("rouge")

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)

    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    result = rouge.compute(
        predictions=decoded_preds,
        references=decoded_labels,
        use_stemmer=True
    )
    # table me percent style (42.85 etc.) chahiye to *100
    return {k: round(v * 100, 6) for k, v in result.items()}
```

```python
class LogTableCallback(TrainerCallback):
    def __init__(self):
        self.rows = []

    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if not metrics:
            return
        self.rows.append({
            "Step": int(state.global_step),
            "Training Loss": metrics.get("train_loss", np.nan),   # kabhi NaN aa sakta
            "Validation Loss": metrics.get("eval_loss", np.nan),
            "Rouge1": metrics.get("eval_rouge1", np.nan),
            "Rouge2": metrics.get("eval_rouge2", np.nan),
            "RougeL": metrics.get("eval_rougeL", np.nan),
            "RougeLsum": metrics.get("eval_rougeLsum", np.nan),
        })

    def to_df(self):
        df = pd.DataFrame(self.rows)
        return df.sort_values("Step").reset_index(drop=True)

log_cb = LogTableCallback()
```

```python
train_result = trainer.train()

print("\nSaving model...")
trainer.save_model("t5_summarizer")
tokenizer.save_pretrained("t5_summarizer")

print("\n" + "="*70)
print(log_cb.to_df().to_string(index=False))
print("="*70)

print("\nTraining Metrics:")
for k, v in train_result.metrics.items():
    print(f"{k}: {v}")
```

```python
import gradio as gr
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

MODEL_DIR = "t5_cnn_dm_final"   # local folder saved above

tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_DIR)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def summarize(text, max_new_tokens=128, num_beams=4):
    if text is None or len(text.strip()) == 0:
        return ""
    inputs = tokenizer(
        "summarize: " + text,
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to(device)

    with torch.no_grad():
        output_ids = model.generate(
            **inputs,
            max_new_tokens=int(max_new_tokens),
            num_beams=int(num_beams),
            early_stopping=True
        )
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)

demo = gr.Interface(
    fn=summarize,
    inputs=[
        gr.Textbox(lines=12, label="Article text"),
        gr.Slider(32, 256, value=128, step=8, label="Max new tokens"),
        gr.Slider(1, 8, value=4, step=1, label="Beams")
    ],
    outputs=gr.Textbox(lines=6, label="Summary"),
    title="T5 CNN/DailyMail Summarizer",
    description="Fine-tuned T5-small summarizer (prefix: 'summarize: ')."
)

demo.launch(share=True)
```
